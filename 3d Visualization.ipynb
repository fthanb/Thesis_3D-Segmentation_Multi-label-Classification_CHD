```python
import nibabel as nib
import numpy as np
import plotly.graph_objects as go
from skimage import measure

mask_path = r"D:/TEKNIK INFORMATIKA/SKRIPSI/Kagayaki/CHD_Classification/Segmentation/fg_Umamba/fold_0/validation/1005.nii.gz"

print("Loading Data")
mask_data = nib.load(mask_path).get_fdata()

colors = {
    1: 'red',       # Left Ventricle
    2: 'blue',      # Right Ventricle
    3: 'green',     # Left Atrium
    4: 'yellow',    # Right Atrium
    5: 'purple',    # Myocardium (Otot Jantung)
    6: 'orange',    # Aorta
    7: 'cyan'       # Pulmonary Artery
}

fig = go.Figure()

print("3D Modelling")

for label_idx, color in colors.items():

    binary_mask = (mask_data == label_idx)

    if not np.any(binary_mask):
        continue

    # Marching Cubes for 3D Surface Mesh
    verts, faces, normals, values = measure.marching_cubes(binary_mask, level=0.5, step_size=2)

    fig.add_trace(go.Mesh3d(
        x=verts[:, 0],
        y=verts[:, 1],
        z=verts[:, 2],
        i=faces[:, 0],
        j=faces[:, 1],
        k=faces[:, 2],
        color=color,
        opacity=0.8,
        name=f'Organ {label_idx}',
        showscale=False
    ))

fig.update_layout(
    title="3D Heart Segmentation (U-Mamba)",
    scene=dict(
        xaxis=dict(visible=False),
        yaxis=dict(visible=False),
        zaxis=dict(visible=False),
        aspectmode='data',
        camera=dict(
            eye=dict(x=1.5, y=1.5, z=1.5)
        )
    ),
    margin=dict(l=0, r=0, b=0, t=40)
)

print("ROTATE")
fig.show()
```

```python
import nibabel as nib
import plotly.graph_objects as go
import numpy as np
from skimage import measure

print("Loading")
original_path = r'D:/TEKNIK INFORMATIKA/SKRIPSI/ImageCHD_dataset/ct_1005_image.nii.gz'
img_data = nib.load(original_path).get_fdata()

threshold_level = 1600

print(f"Threshold: {threshold_level}")
fig = go.Figure()

step_ukuran = 2

try:

    verts, faces, normals, values = measure.marching_cubes(
        img_data,
        level=threshold_level,
        step_size=step_ukuran
    )

    fig.add_trace(go.Mesh3d(
        x=verts[:, 0],
        y=verts[:, 1],
        z=verts[:, 2],
        i=faces[:, 0],
        j=faces[:, 1],
        k=faces[:, 2],
        color='gray',
        opacity=0.3,
        name=f'Original CT (Level: {threshold_level})',
        showscale=False,
        lighting=dict(ambient=0.4, diffuse=0.5, fresnel=0.1, specular=0.5, roughness=0.1),
        lightposition=dict(x=100, y=200, z=150)
    ))

except ValueError as e:
    print(f"Error")

fig.update_layout(
    title=f"3D Original Scan Visualization (Threshold: {threshold_level})",
    scene=dict(
        xaxis=dict(visible=False),
        yaxis=dict(visible=False),
        zaxis=dict(visible=False),
        aspectmode='data',
        camera=dict(
            eye=dict(x=1.5, y=1.5, z=1.5)
        )
    ),
    margin=dict(l=0, r=0, b=0, t=40)
)

print("ROTATE")
fig.show()
```
