## Count Patient Frames

```python
import os
import nibabel as nib
import pandas as pd
from pathlib import Path

dataset_dir = Path(r"D:\TEKNIK INFORMATIKA\SKRIPSI\ImageCHD_dataset")

nifti_files = list(dataset_dir.rglob("ct_*_image.nii.gz"))

print(f"Found {len(nifti_files)} NIfTI files. Processing...")

patient_data = []

for file_path in nifti_files:
    try:
        img = nib.load(file_path)

        shape = img.shape
        
        # The shape is (X, Y, Z) X and Y represent the width and length pixels, also Z represents the number of frames (height for 3D imegs cube).
        total_frames = shape[2] if len(shape) >= 3 else "Unknown"
        
        # Extract Patient ID from the filename (1001 - 1178)
        patient_id = file_path.name.split('_')[1]
   
        patient_data.append({
            "Patient ID": patient_id,
            "Filename": file_path.name,
            "Dimensions (X, Y, Z)": shape,
            "Total Frames": total_frames,
        })
        
    except Exception as e:
        print(f"Error processing {file_path.name}: {e}")

df = pd.DataFrame(patient_data)

df['Patient ID'] = pd.to_numeric(df['Patient ID'])
df = df.sort_values("Patient ID").reset_index(drop=True)

print("\n--- ImageCHD Dataset Summary ---")
print(df.to_string())
```
